```python
# Read all responses for 2019
responses = reader.read_observation_responses(
    pd.Timestamp('2019-03-14T00:00:00'),
    pd.Timestamp('2019-03-15T23:59:59')
)

# Initialize empty lists to store data
data = []
timestamps = []

# Process each response
for response in responses:
    # Get timestamp from the response
    ts = conversion_utils.proto_to_pandas_timestamp(response.timestamp)
    
    # Create a dictionary for this timestamp's measurements
    row_data = {}
    
    # Process each observation in the response
    for obs in response.single_observation_responses:
        if obs.observation_valid:
            # Create column name from device_id and measurement_name
            col_name = f"{obs.single_observation_request.device_id}_{obs.single_observation_request.measurement_name}"
            row_data[col_name] = obs.continuous_value
    
    # Only add rows that have data
    if row_data:
        timestamps.append(ts)
        data.append(row_data)

# Create DataFrame
observation_df = pd.DataFrame(data, index=timestamps)
```

```python
# Read all responses for 2019
action_responses = reader.read_action_responses(
    pd.Timestamp('2019-03-14T00:00:00'),
    pd.Timestamp('2019-03-15T23:59:59')
)

# Initialize empty lists to store data
data = []
timestamps = []

# Process each response
for response in action_responses:
    # Get timestamp from the response
    ts = conversion_utils.proto_to_pandas_timestamp(response.timestamp)
    
    # Create a dictionary for this timestamp's measurements
    row_data = {}
    
    # Process each observation in the response
    for act in response.single_action_responses:
        if act.response_type == 1:
            # Create column name from device_id and measurement_name
            col_name = f"{act.single_action_requests.device_id}_{act.single_action_requests.setpoint_name}"
            row_data[col_name] = act.continuous_value
    
    # Only add rows that have data
    if row_data:
        timestamps.append(ts)
        data.append(row_data)

# Create DataFrame
action_df = pd.DataFrame(data, index=timestamps)
```

```python
# Read reward responses for the same time period
reward_responses = reader.read_reward_responses(
    pd.Timestamp('2019-01-01T00:00:00'),
    pd.Timestamp('2019-12-31T23:59:59')
)

# Count responses without agent_reward_value
total_responses = len(reward_responses)

missing_agent_reward = sum(1 for resp in reward_responses if not resp.agent_reward_value==0)
missing_productivity_reward = sum(1 for resp in reward_responses if not resp.productivity_reward==0)
missing_total_occupancy = sum(1 for resp in reward_responses if not resp.productivity_reward==0)

missing_productivity_regret = sum(1 for resp in reward_responses if not resp.productivity_regret==0)
# Calculate percentages for each metric
percent_missing_agent = (missing_agent_reward / total_responses * 100) if total_responses > 0 else 0
percent_missing_productivity = (missing_productivity_reward / total_responses * 100) if total_responses > 0 else 0
percent_missing_occupancy = (missing_total_occupancy / total_responses * 100) if total_responses > 0 else 0
percent_missing_regret = (missing_productivity_regret / total_responses * 100) if total_responses > 0 else 0

print(f"Total responses: {total_responses}")
print(f"Responses missing agent_reward: {missing_agent_reward} ({percent_missing_agent:.2f}%)")
print(f"Responses missing productivity_reward: {missing_productivity_reward} ({percent_missing_productivity:.2f}%)")
print(f"Responses missing total_occupancy: {missing_total_occupancy} ({percent_missing_occupancy:.2f}%)")
print(f"Responses missing productivity_regret: {missing_productivity_regret} ({percent_missing_regret:.2f}%)")
```
